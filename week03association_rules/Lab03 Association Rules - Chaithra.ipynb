```python
# loading the data set
df = pd.read_csv(r'groceries_one_hot.csv') # path needs to be reaplaced with your file path if running on another machine
```

```python
records = []
for i in range(df.shape[0]):
    records.append([str(df.values[i,j]) for j in range(df.shape[1])])
```

```python
association_rules = apriori(records, min_support=0.0045, min_confidence=0.02, min_lift=1)
association_results = list(association_rules)
```

```python
print(len(association_results))
```

```python
association_results[0]
```

```python
for item in association_rules:

    # first index of the inner list
    # Contains base item and add item
    pair = item[0] 
    items = [x for x in pair]
    print("Rule: " + items[0] + " -> " + items[1])

    #second index of the inner list
    print("Support: " + str(item[1]))

    #third index of the list located at 0th
    #of the third index of the inner list

    print("Confidence: " + str(item[2][0][2]))
    print("Lift: " + str(item[2][0][3]))
    print("=====================================")
```
