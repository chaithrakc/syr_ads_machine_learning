```python
import pandas as pd
```

```python
bankdata=pd.read_csv("bankdata_csv_all.csv")
```

```python
bankdata
```

```python
bankdata.fillna(0)
```

```python
bankdata.set_index('id',inplace=True)
```

```python
bankdata['age'] = pd.cut(bankdata.age,500)
bankdata['income'] =pd.cut(bankdata.income,500)
bankdata['children'] = pd.cut(bankdata.children,500)
```

```python
bankdata.info()
```

```python
bankdata_oneHot = pd.get_dummies(bankdata)
```

```python
pip install mlxtend
```

```python
from mlxtend.frequent_patterns import apriori, fpmax, fpgrowth
from mlxtend.frequent_patterns import association_rules
```

```python
rules = apriori(bankdata_oneHot,min_support = 0.11,use_colnames=True)
```

```python
association_rules(rules, metric="confidence", min_threshold=0.06)
```

```python
association_rules(rules, metric="lift", min_threshold=1)
```

```python
assoc_rules=association_rules(rules, metric="confidence", min_threshold=0.7)
```

```python
assoc_lift=association_rules(rules, metric="lift", min_threshold=1)
```
