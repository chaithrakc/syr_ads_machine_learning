```python
#for plotting images & adjusting colors
import matplotlib.pyplot as plt
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator
import nltk
nltk.download('twitter_samples')
from nltk.corpus import twitter_samples
from nltk.corpus import stopwords
nltk.download('stopwords')
```

```python
import pandas as pd
```

```python
#twitter corpus from nltk
text = twitter_samples.strings('tweets.20150430-223406.json')

# join the list of all the words
text = ' '.join(text)
```

This data comprises of data scraped from rev.com.This is a collection of all Democratic Primary Debate Transcripts up until this point!

```python
twitter_df = pd.read_csv('debate_transcripts_v3_2020-02-26.csv', encoding='cp1252')
twitter_df.head()
```

```python
text = ' '.join(twitter_df['speech'])
```

```python
#create the wordcloud object
# wordcloud = WordCloud(stopwords = STOPWORDS,
#                       collocations=True).generate(text)
def build_cloud(text):
  wordcloud = WordCloud(background_color="white", 
                max_words=350, 
                width=1000, 
                height=600, 
                max_font_size=256,
                random_state=1).generate(text)

  #plot the wordcloud object
  plt.figure(figsize=(15,15))
  plt.imshow(wordcloud)
  plt.axis("off")
  plt.show()
```

```python
build_cloud(text)
```

**DATA PRE-PROCESSING**

```python
# lowercase transforamtion
lowercase_text = text.lower()
build_cloud(lowercase_text)
```

```python
# removing stop words
STOPWORDS = stopwords.words('english')

wordcloud = WordCloud(stopwords = STOPWORDS,
                background_color="white", 
                max_words=350, 
                width=1000, 
                height=600, 
                max_font_size=256,
                random_state=1).generate(text)
```

```python
#plot the wordcloud object
plt.figure(figsize=(15,15))
plt.imshow(wordcloud)
plt.axis("off")
plt.show()
```
